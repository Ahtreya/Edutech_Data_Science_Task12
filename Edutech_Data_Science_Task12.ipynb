```python
import pandas as pd
df=pd.read_csv('/content/Housing.csv')
print(df)
```

```python

# 1. Apply One Hot and Label Encoding
print("---- Applying Label Encoding ----")
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
df['furnishingstatus']=le.fit_transform(df['furnishingstatus'])
print(df['furnishingstatus'])
print("\n")
print("Applying Label Encoding is successful.")
print("\n")
print("---- Applying One Hot Encoding ----")
categorical_col=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
for col in categorical_col:
    df[col]=pd.get_dummies(df[col],drop_first=True,dtype=int)
    print(df[col])
print("\n")
print("Applying One Hot Encoding is successful.")
print("\n")


```

```python
# 2. Create new interaction features
print("---- Creating new interaction features ----")
df['bed_to_bath_ratio']=df['bedrooms']/(df['bathrooms']).clip(lower=1)
df['total_rooms']=df['bedrooms']+df['bathrooms']
df['area_per_room']=df['area']/(df['total_rooms']).clip(lower=1)
print(df['bed_to_bath_ratio'])
print("\n")
print(df['area_per_room'])
print("\n")
print("Creating new interaction features is successful.")
print("\n")
```

```python
# 3. Handle skewed data with log transformation
import numpy as np
print("---- Handling skewed data with log transformation ----")
# Check original skewness values
print(f"Original Price Skewness: {df['price'].skew():.2f}")
print(f"Original Area Skewness: {df['area'].skew():.2f}")
print("-" * 50)
# Apply log transformation
df['log_price']=np.log1p(df['price'])
df['log_area']=np.log1p(df['area'])
# Check new skewness values
print(f"Transformed Price Skewness: {df['log_price'].skew():.2f}")
print(f"Transformed Area Skewness: {df['log_area'].skew():.2f}")
print("-" * 50)
print("Log Transformation is successful.\n")
df.to_csv('EnhancedDataset.csv', index=False)
print("EnhancedDataset.csv has been saved successfully in your workspace!")


```
